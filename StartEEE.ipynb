# See equal prediction

This module aims to find all the equal prediction effectuated by our model. 
We select a team and, for all the frames of the matches analyzed by that model, we look for prediction that are equal both in terms of analyzed and opponent team and for the ball.

We discard from the equal prediction, forecasting that are not at least 2 seconds apart, in order to find how different starting situation lead to the same situation.

```python
from Explain.Encoding_extract_equal import extract_data, main_pred, main_real
```

The team to analyze, write the name of the team as a string. The list of available team is composed by the teams participating in 2019/2020 Serie A.

The module starts to extract equal prediction for the selected team. The process is very slow, but once the process of a team is finished, it saves the dataset and it will be available for the next computation saving time.

```python
#Team to analyze
team ='Juventus'
```

```python
dt, se, equal = extract_data(team)
```

Visualization of the euqla predictions.

With 'equal[i]' we select a specific couple of prediction.

The arrows represent the movements of the players between the frame of closeness, red for the analyzed team and blue for the opponent team, while the yellow arrows represent the ball. The color of each cell acts for the predominance of one  team in that cell, darker is the violet and greater is the number of player of the opponent team with respect of the analyzed team, while darker is the green and greater is the number of player of the analyzed team with respect of the opponent team. The cell with the yellow border shows the cell in which the ball has been predicted.
