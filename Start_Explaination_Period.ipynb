# Explain period

This code aims to understand how a displacement of one or more player will affect the position, in the next 5 seconds, of the two teams involved in the game.

To make movement of player in accordance with the actual disposition of the players, we firstly make a clustering selecting some usual disposition, in order to select the right movement.

Th movement are done on the first frame of period (the closest to the prediction between the five).

```python
from Explain.Explaination_Period import main_explain_per, main_predict
```

# Extract data and cluster

In the variable 'team' just write the team to analyze as a string.

Running the cell, if the dataset is not presented, the module will start to predict all the frames of the matches included for the model.

Then, it will cluster the result and it shows the centroids.

```python
#Team to analyze
team='Roma'

clusters_ = main_explain_per(team)
```

# Change position

Cell to predict how the displacements of one or more players in the first frame of closeness affect the prediction, and thus, the tactics of the team selected.

Variable:
- team_:  0 is the analyzed, 1 the opponent;
- ind: the number of cluster where to make the displacements, the number refers to the cluster shown above;
- row_: the row where the cell from where we have to remove the player is located;
- col_: the col where the cell from where we have to remove the player is located;
- way: could be 'forward', 'behind', 'right', 'left', 'forward-left', 'forward-right', 'behind-left', 'behind-right'. It indicates the direction of the movement of the player.
- value: the number of players to move.

```python
#Team to change position, 0 is the analyzed, 1 the opponent
team_ = 0

#Cluster to choose
ind = 0

#Row and column where we have to remove players
row_ = 4
col_ = 1

#Direction of the movement, it can be:
# 'forward', 'behind', 'right', 'left', 'forward-left', 'forward-right', 'behind-left', 'behind-right'
way = 'forward'

#Number of player to move
value = 1

main_predict(clusters_, ind, team_, row_, col_, way, value, team)
```
