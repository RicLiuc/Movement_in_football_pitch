```python
import numpy as np
import h5py
import params
import matplotlib.pyplot as plt
```

```python
dt = h5py.File("../Files/Juventus/JuventusDataset.h5", 'r')
dt2 = h5py.File("../Files/Spal/SpalDataset.h5", 'r')
```

```python
temp=dt['Y_train']
temp2=dt2['Y_train']
```

```python
ball = [temp[i][2] for i in range(0, len(temp))]
ball2 = [temp2[i][2] for i in range(0, len(temp2))]
```

```python
res = np.zeros((15))
for i in ball:
    for r in range(0, params.row):
        for c in range(0, params.col):
            if i[r][c] != 0:
                res[5*c+r]=res[5*c+r]+1

res2 = np.zeros((15))
for i in ball2:
    for r in range(0, params.row):
        for c in range(0, params.col):
            if i[r][c] != 0:
                res2[5*c+r]=res2[5*c+r]+1
```

```python
firs_third_J = res[0]+res[1]+res[2]+res[3]+res[4]
half_J = res[5]+res[6]+res[7]+res[8]+res[9]
last_third_J = res[10]+res[11]+res[12]+res[13]+res[14]

firs_third_S = res2[0]+res2[1]+res2[2]+res2[3]+res2[4]
half_S = res[5]+res2[6]+res2[7]+res2[8]+res2[9]
last_third_S = res2[10]+res2[11]+res2[12]+res2[13]+res2[14]
```

```python
left_J = res[0]+res[1]+res[5]+res[10]+res[11]
center_J = res[2]+res[6]+res[7]+res[8]+res[12]
right_J = res[3]+res[4]+res[9]+res[13]+res[14]

left_S = res2[0]+res2[1]+res2[5]+res2[10]+res2[11]
center_S = res2[2]+res2[6]+res2[7]+res2[8]+res2[12]
right_S = res2[3]+res2[4]+res2[9]+res2[13]+res2[14]
```

```python
plt.figure(figsize=(20,5))

X = range(0, 15)

Juve = res
Spal = res2
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Juve, 0.4, label = 'Juventus')
plt.bar(X_axis + 0.2, Spal, 0.4, label = 'Spal')
  
plt.xticks(X_axis, X)
plt.xlabel("Cell")
plt.ylabel("Occurence of the ball")
plt.title("Number of occurence of ball in each cell")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

X = range(0, 3)

Juve = [firs_third_J, half_J, last_third_J]
Spal = [firs_third_S, half_S, last_third_S]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Juve, 0.4, label = 'Juventus')
plt.bar(X_axis + 0.2, Spal, 0.4, label = 'Spal')
  
plt.xticks(X_axis, ['Defensive third', 'Middle third', 'Offensive third'])
#plt.xlabel("Zone of the ")
plt.ylabel("Occurence of the ball")
plt.title("Number of occurence of ball in each third")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

X = range(0, 3)

Juve = [left_J, center_J, right_J]
Spal = [left_S, center_S, right_S]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Juve, 0.4, label = 'Juventus')
plt.bar(X_axis + 0.2, Spal, 0.4, label = 'Spal')
  
plt.xticks(X_axis, ['Left side', 'Center', 'Right side'])
#plt.xlabel("Cell")
plt.ylabel("Occurence of the ball")
plt.title("Number of occurence of ball in each cell")
plt.legend()
plt.show()
```
